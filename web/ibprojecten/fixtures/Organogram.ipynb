```python
import requests
import csv
from bs4 import BeautifulSoup
import json
```

```python
#url = 'https://zec.amsterdam.nl/organisatiegids-zoek/'

file = '/Users/eelkejager/Documents/datapunt/projecten/projecten_ib/data/zecorganisaties.html'

with open(file, 'r') as htmlDoc:
    soup = BeautifulSoup(htmlDoc, 'html.parser')
    #See html page
    #print(soup.prettify())
    
    # select all grid-elements and select 4th one with list
    grid = soup.find_all('div', "grid-element")[3]
    #print(grid)
    items = grid.find_all(['h2','a'])
    print(items)
```

```python
    organogram = []
    organisatie ={}
    cluster = ''
    i = 1
    for item in items:   
        organisatie ={"model":'app.organisatie', "fields":{}}
        if item.string is None:
            #print(item.string)
            continue
        if item.name == 'h2':
            #print(item.string)
            if cluster != item.get_text():
               cluster = item.get_text()
        organisatie["fields"]["Cluster"] = cluster
        if item.name == 'a' and item.get_text()[:1] != '.':  
            organisatie["fields"]["Organisatie"]= item.get_text()
            organisatie["org_id"]= i
            organogram.append(organisatie)
            i+=1
    print(organogram)
   
```

```python
with open('organisaties.json', 'w') as outfile:
    json.dump(organogram, outfile,  sort_keys = True, indent = 4,
               ensure_ascii = False)
```
